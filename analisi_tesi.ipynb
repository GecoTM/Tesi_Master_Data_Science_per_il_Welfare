```python
# IEA prevede, in caso di sviluppo accelerato, una crescita del consumo energetico in TWh da parte dei data center di circa il 203% dal 2024 al 2030
## Creo un nuova colonna in cui calcolo l'aumento percentuale
```

```python
# seleziono la colonna '2024'
print(consumo_totale.iloc[:, 3])

# Calcolo l'aumento del consumo energetico del 203% dal 2024 al 2030
consumo_totale['2030'] = (consumo_totale['2024'] * 3.03846154)

# Calcolo l'aumento del 36% previsto per il 2035 (a partire dal 2030) e creo la nuova colonna
consumo_totale['2035'] = (consumo_totale['2030'] * 1.36)

consumo_totale
```

```python
ai_supercomp = pd.read_csv("/Users/jacopo/Desktop/Tesi Master/Dataset_da_analizzare/AI_supercomputers.csv")

ai_supercomp
```

```python
ai_supercomp.columns
```

```python
new_supercomp = ai_supercomp[["Name", "Status", "Owner", "First Operational Date" , "Total number of AI chips"]]

# filtro solo le strutture esistenti e rimuovo le righe vuote del numero totale di chip per l'IA
new_supercomp = new_supercomp[(new_supercomp["Status"] != "Planned") & (new_supercomp["Total number of AI chips"].notna())]

# rimuovo le righe con valori vuoti
new_supercomp = new_supercomp[new_supercomp["First Operational Date"].notna()]

# modifico le righe vuote della colonna "Owner"
new_supercomp["Owner"] = new_supercomp["Owner"].fillna("Not specified")

new_supercomp
```

```python
new_supercomp = new_supercomp.sort_values(by=['First Operational Date'], ascending = True)

new_supercomp
```

```python
new_supercomp = pd.DataFrame(new_supercomp)
```

```python
# converto la colonna 'First Operational Date' in formato data
new_supercomp['First Operational Date'] = pd.to_datetime(new_supercomp['First Operational Date'])

# creo la colonna 'Year' e sommo il numero di chip IA resi operativi per ogni anno 
new_supercomp['Year'] = new_supercomp['First Operational Date'].dt.year
chips_per_year = new_supercomp.groupby('Year')['Total number of AI chips'].sum()
```

```python
chips_per_year
```

```python
chips_per_year.to_csv("chips_per_year.csv")
```
